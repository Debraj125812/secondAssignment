```python
import csv
from collections import Counter
import re
```

```python
def tokenize_text(text):
    # Remove any non-alphabetic characters and split the text into words
    words = re.findall(r'\b\w+\b', text.lower())  # Convert to lowercase for case-insensitive counting
    return words
```

```python
text_file = r"C:\Users\DELL\Desktop\Assignemnt2\combined_texts.txt"
```

```python
with open(text_file, 'r', encoding='utf-8') as file:
    text_content = file.read()

# Tokenize the text
words = tokenize_text(text_content)

# Count occurrences of each word
word_counts = Counter(words)

# Get the 30 most common words
top_30_words = word_counts.most_common(30)

# Save the Top 30 words and their counts to a CSV file
output_csv = 'top_30_words.csv'
with open(output_csv, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Word', 'Count'])  # Write the header
    writer.writerows(top_30_words)  # Write the word and their counts

print(f"Top 30 words have been saved to {output_csv}.")
```

```python
from collections import Counter
from transformers import AutoTokenizer

# Function to tokenize and count unique tokens
def count_top_tokens(text_file, model_name="bert-base-uncased", top_n=30):
    # Load the AutoTokenizer from the Transformers library
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    # Read the text file
    with open(text_file, 'r', encoding='utf-8') as file:
        text = file.read()

    # Tokenize the text
    tokens = tokenizer.tokenize(text)

    # Count occurrences of each token
    token_counts = Counter(tokens)

    # Get the top N tokens
    top_tokens = token_counts.most_common(top_n)

    # Print the top tokens
    print(f"Top {top_n} tokens:")
    for token, count in top_tokens:
        print(f"{token}: {count}")

    return top_tokens

# Example usage
text_file = r"C:\Users\DELL\Desktop\Assignemnt2\combined_texts.txt"
top_30_tokens = count_top_tokens(text_file, model_name="bert-base-uncased", top_n=30)

# To store the result in a CSV (optional)
import csv

# Save the Top 30 tokens and their counts to a CSV file
output_csv = 'top_30_tokens.csv'
with open(output_csv, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Token', 'Count'])  # Write the header
    writer.writerows(top_30_tokens)  # Write the tokens and their counts

print(f"Top 30 tokens have been saved to {output_csv}.")
```

```python
import spacy
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
from collections import Counter

# Load the scispaCy models for detecting diseases and drugs
nlp_bc5cdr = spacy.load("en_ner_bc5cdr_md")

# Load the BioBERT model for NER
biobert_model = "dmis-lab/biobert-base-cased-v1.1"
tokenizer = AutoTokenizer.from_pretrained(biobert_model)
model = AutoModelForTokenClassification.from_pretrained(biobert_model)
biobert_ner = pipeline("ner", model=model, tokenizer=tokenizer)

# Function to read the text from the file
def read_text_file(text_file):
    with open(text_file, 'r', encoding='utf-8') as file:
        return file.read()

# Function to extract diseases and drugs using scispaCy (BC5CDR)
def extract_entities_bc5cdr(text):
    doc = nlp_bc5cdr(text)
    diseases = [ent.text for ent in doc.ents if ent.label_ == 'DISEASE']
    drugs = [ent.text for ent in doc.ents if ent.label_ == 'CHEMICAL']
    return diseases, drugs

# Function to extract diseases and drugs using BioBERT
def extract_entities_biobert(text):
    ner_results = biobert_ner(text)
    diseases = []
    drugs = []
    for entity in ner_results:
        if "DISEASE" in entity['entity']:
            diseases.append(entity['word'])
        elif "CHEMICAL" in entity['entity']:
            drugs.append(entity['word'])
    return diseases, drugs

# Function to compare the entities extracted by both models
def compare_entities(diseases_bc5cdr, drugs_bc5cdr, diseases_biobert, drugs_biobert):
    diseases_bc5cdr_count = Counter(diseases_bc5cdr)
    drugs_bc5cdr_count = Counter(drugs_bc5cdr)
    diseases_biobert_count = Counter(diseases_biobert)
    drugs_biobert_count = Counter(drugs_biobert)

    print("\nTotal Disease Entities (BC5CDR):", len(diseases_bc5cdr))
    print("Total Drug Entities (BC5CDR):", len(drugs_bc5cdr))
    print("Total Disease Entities (BioBERT):", len(diseases_biobert))
    print("Total Drug Entities (BioBERT):", len(drugs_biobert))

    # Common and unique entities
    common_diseases = set(diseases_bc5cdr).intersection(set(diseases_biobert))
    common_drugs = set(drugs_bc5cdr).intersection(set(drugs_biobert))
    unique_diseases_bc5cdr = set(diseases_bc5cdr).difference(set(diseases_biobert))
    unique_drugs_bc5cdr = set(drugs_bc5cdr).difference(set(drugs_biobert))
    unique_diseases_biobert = set(diseases_biobert).difference(set(diseases_bc5cdr))
    unique_drugs_biobert = set(drugs_biobert).difference(set(drugs_bc5cdr))

    print("\nCommon Disease Entities:", len(common_diseases))
    print("Common Drug Entities:", len(common_drugs))
    print("\nUnique Disease Entities (BC5CDR):", len(unique_diseases_bc5cdr))
    print("Unique Drug Entities (BC5CDR):", len(unique_drugs_bc5cdr))
    print("Unique Disease Entities (BioBERT):", len(unique_diseases_biobert))
    print("Unique Drug Entities (BioBERT):", len(unique_drugs_biobert))

    # Most common entities
    print("\nMost common diseases in BC5CDR:", diseases_bc5cdr_count.most_common(5))
    print("Most common drugs in BC5CDR:", drugs_bc5cdr_count.most_common(5))
    print("\nMost common diseases in BioBERT:", diseases_biobert_count.most_common(5))
    print("Most common drugs in BioBERT:", drugs_biobert_count.most_common(5))

# Main function to run the comparison
def main(text_file):
    text = read_text_file(text_file)

    # Extract entities using scispaCy (BC5CDR)
    diseases_bc5cdr, drugs_bc5cdr = extract_entities_bc5cdr(text)

    # Extract entities using BioBERT
    diseases_biobert, drugs_biobert = extract_entities_biobert(text)

    # Compare the entities extracted by both models
    compare_entities(diseases_bc5cdr, drugs_bc5cdr, diseases_biobert, drugs_biobert)

# Example usage
text_file = r"C:\Users\DELL\Desktop\Assignemnt2\combined_texts.txt"  # Replace with the path to your text file
main(text_file)
```
